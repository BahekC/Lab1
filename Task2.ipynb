```python
from typing import List

def text_to_words(text: str) -> List[str]:
    """
    Превращает текст в список слов согласно условию:
    - слово = последовательность букв (isalpha)
    - пробелы, переносы, знаки препинания и прочие символы разрывают слово
    - все небуквенные символы игнорируются (фактически работают как разделители)
    - слова приводим к нижнему регистру
    - порядок слов сохраняется
    """
    words = []
    buf = []  # сюда накапливаем буквы текущего слова

    for ch in text:
        if ch.isalpha():          # буква -> часть слова
            buf.append(ch.lower())
        else:
            # встретили разделитель -> закрываем текущее слово (если оно есть)
            if buf:
                words.append("".join(buf))
                buf.clear()

    # если текст закончился на буквах — добавим последнее слово
    if buf:
        words.append("".join(buf))

    return words


# ===== ГОТОВЫЙ ПРИМЕР =====
sample_text = "We study programming languages C++, C#, Go. We are programmers!"
words = text_to_words(sample_text)

print("Текст:", sample_text)
print("Список слов:", words)
```

```python
from collections import defaultdict
from typing import Dict

def build_bigrams(words: List[str]) -> Dict[str, List[str]]:
    """
    Строит словарь биграмм:
    ключ = первое слово биграммы
    значение = список слов, которые встречаются после ключа (с сохранением повторов)
    """
    bigrams = defaultdict(list)
    for i in range(len(words) - 1):
        first = words[i]
        second = words[i + 1]
        bigrams[first].append(second)
    return dict(bigrams)


# ===== ГОТОВЫЙ ПРИМЕР НА sample_text =====
bigrams = build_bigrams(words)

print("Словарь биграмм (пример):")
for k in sorted(bigrams.keys()):
    print(f"  {k!r}: {bigrams[k]}")
```

```python
import random
from typing import Optional

def autocomplete(seed: str, bigrams: Dict[str, List[str]], add_words: int = 3, *, rng: Optional[random.Random] = None) -> str:
    """
    seed: стартовое слово
    add_words: сколько слов дописать (3-4 по заданию)
    rng: генератор случайных чисел (для воспроизводимости можно передать random.Random(0))
    """
    if rng is None:
        rng = random

    seed = seed.lower()

    if seed not in bigrams:
        return "Слово не найдено"

    result = [seed]
    current = seed

    for _ in range(add_words):
        if current not in bigrams:   # цепочка оборвалась
            break
        next_word = rng.choice(bigrams[current])  # случайное продолжение :contentReference[oaicite:3]{index=3}
        result.append(next_word)
        current = next_word

    return " ".join(result)


# ===== ГОТОВЫЙ ПРИМЕР ВВОДА =====
# (фиксируем seed для воспроизводимости)
rng = random.Random(42)

queries = ["we", "languages", "python"]

print("Запросы:", queries)
print("Ответы:")
for q in queries:
    print(" ", autocomplete(q, bigrams, add_words=3, rng=rng))
```

```python
# ===== ОБУЧЕНИЕ НА ТЕКСТЕ ИЗ ФАЙЛА =====
path = "input.txt"  # поменяйте на ваш файл
with open(path, "r", encoding="utf-8") as f:
    text = f.read()

words = text_to_words(text)
bigrams = build_bigrams(words)

print("Обучение завершено.")
print("Всего слов (с повторами):", len(words))
print("Уникальных ключей (первых слов биграмм):", len(bigrams))


# ===== ИНТЕРАКТИВНЫЙ РЕЖИМ =====
# вводите слово, программа дописывает 3-4 слова; 'exit' для выхода
rng = random.Random()  # можно random.Random(0) для воспроизводимости

while True:
    seed = input("Введите слово (или exit): ").strip()
    if not seed or seed.lower() == "exit":
        break
    print(autocomplete(seed, bigrams, add_words=4, rng=rng))
```
