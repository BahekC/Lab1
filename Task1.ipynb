```python
def multiset_intersection(A, B):
    """
    Задание 1
    Пересечение мультимножеств (списков) A и B.
    Каждый элемент x входит min(count_A[x], count_B[x]) раз.
    Порядок результата произвольный.
    """
    if len(A) <= len(B):
        small, large = A, B
    else:
        small, large = B, A

    counts = {}
    for x in small:
        counts[x] = counts.get(x, 0) + 1

    result = []
    for x in large:
        c = counts.get(x, 0)
        if c > 0:
            result.append(x)
            if c == 1:
                del counts[x]
            else:
                counts[x] = c - 1

    return result


# ===== ГОТОВЫЙ ПРИМЕР ВВОДА (A и B) =====
A = [1, 2, 2, 3, 3, 3, 10]
B = [2, 2, 2, 3, 3, 4, 10, 10]

# ===== ВЫЧИСЛЕНИЕ И ВЫВОД =====
C = multiset_intersection(A, B)

print("A =", A)
print("B =", B)
print("Пересечение (C) =", C)
```
